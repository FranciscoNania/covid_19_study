## Section 2 - Data Understading

##### It's import we understand how this dataset works. It has every entry from every city in Brazil since INSERIRDATA regarding covid19, it also has the same information for each state. So most numbers are duplicated

##### For that reason, I decided to split the dataset into two, states and cities. We will also do different studies for each

```python
#Split the dataframe into cities and states. We will work with both of them later on
df_state = df[df['place_type'] == 'state']
df_city = df[df['place_type'] == 'city']
```

##### The first graphic we want to see, is the pandemic behavior on the entire country.

```python
#df_brazil is a dataframe with only the information relevant to create a graphic behavior
df_brazil = df_state.groupby(['epidemiological_week'], as_index=False, sort=False)['new_deaths'].sum()
df_brazil['epidemiological_week'] = df_brazil['epidemiological_week'].astype(str)
```

```python
def plot_behavior(df, location, save_png = False, size = (18,9)):
    '''
    Funtion to print a plot of the covid19 behavior from a dataframe
    
    INPUT:
    df - dataframe
    size - must be a touple with two itens ()
    location - Where the behavior is being studied
    save_png - Instead of plotting the graphic, save it in a .png file. This is the way I used to print plots side by side
    '''
    
    
    #pd.options.plotting.backend
    ax = df.plot(x = 'epidemiological_week', y = 'new_deaths', kind = 'bar', figsize = size, rot = 45)
    ax.set_ylabel('New deaths')
    ax.set_xlabel('Epidemiological Week')
    ax.set_title('Covid 19 behavior - '+ location)
    ax.xaxis.set_major_locator(plt.MaxNLocator(15))
    
    if save_png == True:
        plt.tight_layout()
        plt.savefig('plots/plot'+location+'.png')
        plt.close()
plot_behavior(df_brazil, 'Brazil', save_png = False)   
```

##### From this plot, we can reach a few conclusions, there were a first and a second wave, 
##### the vaccination might have had a significant influence on the behavior, but we will
##### look into that later on. It is also important to see if this behavior is similiar to 
##### each region, so we can start looking for a pattern.

```python
#The visualization of 27 plots is quite complicated, so I chose to save the files and load them side by side for a neat presentation
states = df_state['state'].value_counts().index.tolist()

  
states = df_state['state'].value_counts().index.tolist()
for state in states:
    df_single_state = df_state[df_state['state'] == state]
    df_single_state = df_single_state.groupby(['epidemiological_week'], as_index=False, sort=False)['new_deaths'].sum()
    df_single_state['epidemiological_week'] = df_single_state['epidemiological_week'].astype(str)
    plot_behavior(df_single_state, state, save_png = True, size = (5,3))
```
